```python
#: gutenberg transformation

book_data = db["gutenberg"].find({})
book_df = pd.DataFrame(book_data)
```

```python
book_df = book_df.rename(columns={"Title": "title"})
```

```python
book_df["specifications"] = book_df.drop(columns=["_id", "url", "title", "image_url"]).apply(lambda row: row.to_dict(), axis=1)
book_specifications_df = book_df[["url", "title", "image_url", "specifications"]]
```

```python
book_specifications_df["product_type"] = "book"
```

```python
book_specifications_df
```

```python
books_records = book_specifications_df.to_dict(orient="records")
from pymongo import UpdateOne

queries = [
    UpdateOne(
        {"url": row["url"]},
        {"$set": row},
        upsert=True
    ) for row in books_records
]

for number in range(0, len(queries) - 1, 1000):
    db["products"].bulk_write(queries[number:min(number + 1000, len(queries))])
```
