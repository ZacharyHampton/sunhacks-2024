```python
from dotenv import load_dotenv
import os
import requests
import json
import pymongo
from pymongo import UpdateOne
import pandas as pd
```

```python
load_dotenv()
client = pymongo.MongoClient(os.getenv("MONGODB_URI"))
```

```python
df = pd.read_csv("gutenberg.csv")
```

```python
df
```

```python
df["url"] = df["Text#"].apply(lambda x: "https://www.gutenberg.org/ebooks/" + str(x))
df["image_url"] = df["Text#"].apply(lambda x: "https://www.gutenberg.org/cache/epub/{}/pg{}.cover.medium.jpg".format(x, x))
```

```python
df
```

```python
queries = [UpdateOne(
    {"Text#": record["Text#"]},
    {"$set": record},
    upsert=True
) for record in df.to_dict(orient="records")]

client["product_db"]["gutenberg"].bulk_write(queries)
```
